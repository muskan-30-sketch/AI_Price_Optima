```python
import pandas as pd

df = pd.read_csv("raw/Amazon.csv")  # change name if needed
df.head()
```

```python
df.columns.tolist()
```

```python
sales_columns = [
    "OrderID",
    "OrderDate",
    "CustomerID",
    "ProductID",
    "Quantity",
    "UnitPrice",
    "Discount",
    "Tax",
    "ShippingCost",
    "TotalAmount",
    "PaymentMethod",
    "OrderStatus",
    "City",
    "State",
    "Country"
]

inventory_columns = [
    "ProductID",
    "ProductName",
    "Category",
    "Brand",
    "SellerID"
]

sales_df = df[sales_columns].copy()
inventory_df = df[inventory_columns].copy()

sales_df.head(), inventory_df.head()
```

```python
sales_df = sales_df.sample(n=min(50, len(sales_df)), random_state=42)
inventory_df = inventory_df.sample(n=min(50, len(inventory_df)), random_state=42)
```

```python
sales_df.to_csv("raw/sales_data.csv", index=False)
inventory_df.to_csv("raw/inventory_data.csv", index=False)

print("✅ sales_data.csv and inventory_data.csv created successfully!")
```

```python
inventory_df.to_csv("raw/inventory_data.csv", index=False)
print("inventory_data.csv recreated")
```
